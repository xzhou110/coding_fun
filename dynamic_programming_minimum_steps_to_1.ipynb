Identify minimum number of steps to reach 1 for a given number. Available operations to use:
-  Subtract 1
-  Divide by 2
-  Divide by 3

If the given number is one, minimum number of step is zero.

```python
def min_steps(n):
    '''
    Top down approach without memoization
    type n: int
    rtype: int
    
    '''
    if n<1:
        return 'Not valid. Input must be an interger greater than or equal to 1'
    if n==1:
        return 0
    
    num_steps = min_steps(n-1)    
    if n%2 ==0:
        num_steps = min (num_steps, min_steps(n/2) )
    if n%3 ==0:
        num_steps = min (num_steps, min_steps(n/3) )
    
    
    return num_steps+1
```

```python
min_steps(300)
```

```python
dic = {}


def min_steps2(n):
    '''
    Top down approach with memoization
    Use dictionary or cache to store result for efficiency purpose
    type n: int
    rtype: int
    
    '''
    # Retrieve the steps for the number from cache if already exsits
    if dic.get(n):
        return dic.get(n)
    else:
    
        if n<1:
            return 'Not valid. Input must be an interger greater than or equal to 1'
        if n==1:
            return 0

        num_steps = min_steps(n-1)    
        if n%2 ==0:
            num_steps = min (min_steps(n-1), min_steps(n/2) )
        if n%3 ==0:
            num_steps = min (num_steps, min_steps(n/3) )
            
    # Store the steps for the current number
        dic[n] = num_steps+1
        return num_steps+1
```

```python
def min_steps2b(n):
    
    '''
    Top down approach with build in dictionary
    
    '''
    if n<1:
        return 'Error'
    else:
        dic = {}
        dic[1] = 0
        if n==1:
            return dic[1]
        else:
            return min_steps2b_dic(n, dic)
        
def min_steps2b_dic(n, dic):
    if n in dic.keys():
        return dic[n]
    else:
        min_steps = min_steps2b_dic(n-1,dic)+1
        if n%2==0:
            min_steps = min(min_steps, min_steps2b_dic(n/2, dic)+1)
        if n%3==0:
            min_steps = min(min_steps, min_steps2b_dic(n/3, dic)+1)
        
        dic[n] = min_steps
        return dic[n]
```

```python
def min_steps3(n):
    '''
    Bottom up approach with memoization
    Use dictionary or cache to store result for efficiency purpose
    Instead of recurision, use dictionary to achieve bottom-up goal.
    type n: int
    rtype: int
    
    '''
    # Retrieve the steps for the number from cache if already exsits
    
    
    if n<1:
        return 'Not valid. Input must be an interger greater than or equal to 1'
    
    dic = {}
    for i in range (1, n+1):
        if i == 1:
            dic[1] = 0
        else:
            dic[i] = dic[i-1]    
            if i%2 ==0:
                dic[i] = min (dic[i], dic[i/2] )
            if i%3 ==0:
                dic[i] = min (dic[i], dic[i/3] )

        # Store the steps for the number from cache if already exsits
            dic[i] = dic[i]+1
    
    print (dic[n], '\n\n', dic)
    return dic[n]
```
